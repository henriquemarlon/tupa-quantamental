```python
# imports necessary libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler,MinMaxScaler
from sklearn.metrics import mean_squared_error,make_scorer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```python
df = pd.read_csv("../data/final/final2.csv")
```

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
df = df.drop("Date",axis=1)
df = df.drop("change",axis=1)
df = df.drop("ena",axis=1)
df = df.drop("Fluxo",axis=1)
df = df.drop("DT_COMPTC",axis=1)
```

```python
df = df.tail(len(df))
```

```python
df 
```

```python
# #características
# X_train = df.drop(["Price"],axis=1)[:2800]
# X_test = df.drop(["Price"],axis=1)[2800:]
X = df.shift(-1).drop(["Price"],axis=1)
y  = df["Price"]
X =X.dropna()
y = y[:801]

# train_size = 0.05684
# train_end = int(len(y)*train_size)

# y_train = df[["Price"]][:train_end]
# y_test = df[["Price"]][train_end:]

# X_train = df.drop("Price",axis=1)[:train_end]
# X_test = df.drop("Price",axis=1)[train_end:]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05684, random_state=42)
```

```python
X
```

```python
y
```

```python
model = MLPRegressor(activation= "identity",
 alpha= 0.001,
 early_stopping= True,
 hidden_layer_sizes= (50, 20),
 shuffle= True,
 solver= "lbfgs")
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
