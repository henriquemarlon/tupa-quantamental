```python
import pandas as pd
import os
import numpy as np
```

```python
def fn(path):       # 1.Get file names from directory
    file_list=os.listdir(path)
    file_list.remove("readme.txt")
    return file_list
```

```python
df = pd.read_csv("../../raw_data/ENA/"+fn("../../raw_data/ENA")[0],sep=";")
```

```python
df
```

```python
peso = {
    "N":0.07092,
    "NE":0.17021,
    "S":0.17730,
    "SE":0.58155
}
ag = pd.DataFrame({})
for file in fn("../../raw_data/ENA"):
    print(file)
    df = pd.read_csv("../../raw_data/ENA/"+file,sep=";")
    df = df[["id_subsistema","ena_data","ena_armazenavel_res_mwmed"]]
    df["ena_data"] = pd.to_datetime(df["ena_data"])
    df["weight"] = 0

    df["weight"] = df["id_subsistema"].apply(lambda x:peso[x])
    df["ena"] = df["weight"]*df["ena_armazenavel_res_mwmed"]
    df2 = df[["ena_data","ena"]].groupby("ena_data").sum()
    if list(ag.columns):
        
        ag =  pd.concat([ag,df2])
        
    else:
        ag = df2
```
